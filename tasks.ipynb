## Tasks
*	Range of date from 1860 to 1885
*	Horizontal green solid marker at Y=4
*	Thick horizontal line marker at Y=6
*	Display ‘Time series’ as title
*	Display ‘Number of discoveries’ on the Y axis
*	Display ‘Date’ on the X axis


### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load Dataset

```python
df2 = pd.read_csv(r'Data/ch1_discoveries.csv')
df2["date"] = pd.to_datetime(df2["date"])
df2 = df2.set_index('date')
df2.plot()
```

### Plotting

```python
plt.style.use('ggplot')
df2_subset = df2['01-01-1860':'01-01-1885']
df2_subset.plot(color='blue', fontsize=12, linestyle='--')
plt.title("Time Series", fontsize=14)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Number of discoveries", fontsize=14)
plt.legend(loc="upper left")
plt.axhline(y=4, color='r')
plt.axhline(y=6, color='g', linewidth=3)
plt.show()
```
