```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
```

### Read Dataset 

```python
df = pd.read_csv(r'Data/ch2_co2_levels.csv')

df.columns
```

### Data Types 

```python
print(df.dtypes)
```

### Preview Data 

#### First 5 rows

```python
df.head(5)
```

#### Last 5 Rows

```python
df.tail(5)
```

### Convert `datestamp` column to DateTime 

```python
df["datestamp"] = pd.to_datetime(df["datestamp"], dayfirst=False)
```

### Plotting Time Series Data

```python
df = df.set_index("datestamp")
df.plot()
plt.show()  
```

### Add Styles 

```python
plt.style.use('fivethirtyeight')
df.plot()
plt.show()
```

### Add Description


```python
df.plot(color="blue")

plt.xlabel("Date")
plt.ylabel("CO2")
plt.title("CO2 Emissions")
plt.show()
```

### Adding Some Styles 


```python
df.plot(figsize(20, 15), fontsize=12, linewidth=3, linestyle="--")
plt.xlabel("Date", fontsize=16)
plt.ylabel("CO2", fontsize=16)
plt.title("CO2 Emissions", fontsize=16)
plt.legend(loc="upper right")
plt.show()
```
